```python
ruta_output_steam = '/content/drive/MyDrive/TP N1 Henry/conversion datos/output_steam_games.json'
ruta_aust_items = '/content/drive/MyDrive/TP N1 Henry/conversion datos/australian_users_items.json'
ruta_aust_reviews = '/content/drive/MyDrive/TP N1 Henry/conversion datos/australian_user_reviews.json'
# Lista para almacenar los objetos JSON
lista_json_steam = []
lista_json_items = []
lista_json_reviews = []
# Carga el archivo JSON línea por línea
with open(ruta_output_steam, 'r') as file:
    for line in file:
        try:
            obj = json.loads(line)
            lista_json_steam.append(obj)
        except json.JSONDecodeError as e:
            print(f"Error de decodificación en la línea: {line}")
            print(e)

# Abrir el archivo JSON en modo lectura
with open(ruta_aust_items, 'r', encoding='utf-8') as f_reviews:
    # Iterar sobre cada línea en el archivo
    for line in f_reviews:
        # Utilizar ast.literal_eval para evaluar la línea como un diccionario de Python
        # Luego, usar json.dumps para convertir el diccionario a una cadena JSON
        # Finalmente, usar json.loads para cargar la cadena JSON en un diccionario
        lista_json_items.append(json.loads(json.dumps(ast.literal_eval(line))))
```

```python
# Abrir el archivo JSON en modo lectura
with open(ruta_aust_reviews, 'r', encoding='utf-8') as f_reviews:
    # Iterar sobre cada línea en el archivo
    for line in f_reviews:
        # Utilizar ast.literal_eval para evaluar la línea como un diccionario de Python
        # Luego, usar json.dumps para convertir el diccionario a una cadena JSON
        # Finalmente, usar json.loads para cargar la cadena JSON en un diccionario
        lista_json_reviews.append(json.loads(json.dumps(ast.literal_eval(line))))
```

```python
pd_output_steam = pd.DataFrame(lista_json_steam)
pd_output_steam =pd_output_steam.dropna(how='all')
pd_output_steam.head()
```
